```python
import requests
import xml.etree.ElementTree as ET

# API End Point
url = 'http://apis.data.go.kr/9710000/BillInfoService2/getBillInfoList'

# 요청 변수 및 인증키 설정
params = {
    'serviceKey': 'UJY+e286zOQsAHMHd/5cggpYFaFqG5mWawJKgrubJeKRBqVp0VUsyeHIgw/VGPQjWRSp6yaR/sUhXlhpKyv1cg==',
    'pageNo': '1',
    'numOfRows': '10',
}

# API 호출
response = requests.get(url, params=params)

# 응답 XML에서 특수문자 제거
xml_content = response.text.replace('&#xD;', '')

# XML 응답 정리 및 파싱
root = ET.fromstring(xml_content)


# XML 파일로 저장
with open("실험용10개 추출.xml", "w", encoding="utf-8") as f:
    f.write(xml_content)
```

```python
import xml.etree.ElementTree as ET
import pandas as pd

# XML 파일 읽기
tree = ET.parse('실험용10개 추출.xml')
root = tree.getroot()

# 각 태그의 내용을 추출
data = []
for item in root.findall(".//item"):
    bill_name = item.find("billName").text if item.find("billName") is not None else 'None'
    bill_no = item.find("billNo").text if item.find("billNo") is not None else 'None'
    summary = item.find("summary").text if item.find("summary") is not None else 'None'
    propose_dt = item.find("proposeDt").text if item.find("proposeDt") is not None else 'None'  # 제안일자 추출

    data.append([bill_name, bill_no, summary, propose_dt])

# 데이터프레임 생성
df = pd.DataFrame(data, columns=["의안명", "의안번호", "주요내용", "제안일자"])

# UTF-8 BOM으로 파일 열기
with open("extracted_data.csv", "w", encoding="utf-8-sig", newline="") as f:
    df.to_csv(f, index=False)
```

```python
import requests
import xml.etree.ElementTree as ET
import pandas as pd

base_url = 'http://apis.data.go.kr/9710000/BillInfoService2/getBillInfoList'

# 기본 파라미터 설정
params = {
    'serviceKey': 'UJY+e286zOQsAHMHd/5cggpYFaFqG5mWawJKgrubJeKRBqVp0VUsyeHIgw/VGPQjWRSp6yaR/sUhXlhpKyv1cg==',
    'numOfRows': '10',
    'start_propose_date': '2015-09-01',  #시작제안일자
}

# 결과를 저장할 리스트 초기화
data = []

# pageNo를 1부터 10,000까지 변화시키며 데이터 추출
for pageNo in range(1, 10):
    params['pageNo'] = str(pageNo)

    # API 호출
    response = requests.get(base_url, params=params)
    root = ET.fromstring(response.text)

    for item in root.findall(".//item"):
        bill_name = item.find("billName").text if item.find("billName") is not None else 'None'
        bill_no = item.find("billNo").text if item.find("billNo") is not None else 'None'
        summary = item.find("summary").text if item.find("summary") is not None else 'None'
        data.append([bill_name, bill_no, summary])

    # Optional: 너무 많은 요청을 한번에 보내지 않도록 간격을 두고 요청할 수도 있습니다.
    # time.sleep(1)

# 데이터프레임 생성
df = pd.DataFrame(data, columns=["의안명", "의안번호", "주요내용"])

# UTF-8 BOM으로 파일 열기
with open("for루프 실험.csv", "w", encoding="utf-8-sig", newline="") as f:
    df.to_csv(f, index=False)
```
